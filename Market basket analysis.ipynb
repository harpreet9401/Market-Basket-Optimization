```python
data = df.copy()
```

```python
data1 = data.copy()
```

```python
df.head()
```

```python
data = pd.get_dummies(data['itemDescription'])
```

```python
data.head()
```

```python
data1.drop(['itemDescription'],axis=1,inplace=True)
```

```python
data1
```

```python
data1 = data1.join(data)
```

```python
data1
```

```python
products= df['itemDescription'].unique()
```

```python
products
```

```python
len(products)
```

```python
data2 = data1.groupby(['Member_number','Date'])[products[:]].sum()
```

```python
data2.head()
```

```python
data2 = data2.reset_index()[products]
```

```python
data2
```

```python
data2.values
```

```python
def funct(data):
    for i in products:
        if data[i]>0:
            data[i] = i
    return data
```

```python
data2 = data2.apply(funct,axis=1)
```

```python
data2
```

```python
new_data = data2.values
```

```python
new_data
```

```python
new_data = [i[i!=0].tolist() for i in new_data if i[i!=0].tolist()]
```

```python
new_data
```

```python
association = apriori(new_data,min_support=0.0003,min_confidence=0.05, max_length=5,min_lift=3)
```

```python
result = list(association)
```
